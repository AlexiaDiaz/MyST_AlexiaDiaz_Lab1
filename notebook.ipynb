## <font color= #6B6B6B> 3.1 Libro de ordenes </font>

Un libro de rdenes es una lista de operaciones, ya sean electrnicas o manuales, que una bolsa utiliza para registrar el inters del mercado en un valor especfico o instrumento financiero en un timestamp* especfico. 
Las acciones se cotizan normalmente en un libro de rdenes por volumen y por nivel de precios. 

*marca temporal, conocida tambin como registro de tiempo, cronomarcador, sello de tiempo o timestamp, es una secuencia de caracteres que denotan la hora y fecha en la/s que ocurri determinado evento
